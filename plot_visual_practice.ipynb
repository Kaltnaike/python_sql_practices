```python
#Days of the week
days = [0,1,2,3,4,5,6]
#Your money spent on lunch
money_spent = [50,190,200,250,600,60,90]
#Your friends's money spent on lunch
money_spent1 = [70,30,45,150,180,175,300] 
#Plot your money
plt.plot(days,money_spent)
#Plot your friend's money
plt.plot(days,money_spent1)
#Display the result
plt.show()
```

They are two lines on the same axes. By default the first line is always blue and the next line is always orange but it can be customized.


```python
#line styles
#customizing line colors
plt.plot(days,money_spent, color='green')
plt.plot(days,money_spent1, color='yellow')
plt.show()
```

```python
#Change the style of the plotted line
#Dashed style
plt.plot(days,money_spent, color='green', linestyle='--')
plt.show()
```

```python
#Change the style of the plotted line
#Dotted style
plt.plot(days,money_spent1, color='yellow', linestyle=':')
plt.show()
```

```python
#Change the style of the plotted line
#No line style
plt.plot(days,money_spent, color='green', linestyle='')
plt.show()
```

```python
#point markers-use a keyword argument marker  to emphasize a point  with a specified marker.
#A circle
plt.plot(days,money_spent, color='green', marker='o')
plt.show()
```

```python
#A square
plt.plot(days,money_spent, color='green', marker='s')
plt.show()
```

```python
#A star
plt.plot(days,money_spent, color='green', marker='x')
plt.show()
```

```python
#Labelling axes
plt.plot(days,money_spent)
plt.plot(days,money_spent1)
plt.xlabel("Days of the week")
plt.ylabel("Money spent on lunch")
plt.title("Lunch money during the week between me and tife")
plt.show()
```
