```python
# Subset the housing data itself
house_data_subset = df[['price', 'bedrooms', 'bathrooms', 'city', 'country']]

house_data_subset.head()
```

```python
# the groupby object
group_by_city = house_data_subset.groupby('city')
```

```python
type(group_by_city)
```

```python
group_by_city.first()
```

```python
group_by_city.last()
```

```python
seattle = group_by_city.get_group('Seattle')
```

```python
#get the average house price for seattle only
seattle['price'].mean()
```

```python
# Get average house price for all cities in the dataset
group_by_city['price'].mean()
```

```python
# get maximum bedrooms
group_by_city['bedrooms'].max()
```

```python
# get the most expensive houses for each city
max_house_prices = house_data_subset.groupby('city').price.max().reset_index()
```

```python
# The most expensive house in seattle city
max_house_prices[max_house_prices['city'] == 'Seattle']
```

```python
expensive_houses = house_data_subset.groupby('city').price\
                    .apply(lambda x: np.percentile(x, 75))\
                    .reset_index()
expensive_houses
```
